```python
import pandas as pd
df = dataset["train"].to_pandas()
```

```python
df.head()
df.info()
df.describe()
```

```python
# Vérifier s'il y a des doublons dans tout le dataset
df.duplicated().sum()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df.duplicated().sum()
```

```python
#les valeurs Manquantes
df.dropna(inplace=True)
```

```python
# Supprimer les espaces au début/fin et mettre les noms en minuscule ou majuscule cohérente
df['Medicine Name'] = df['Medicine Name'].str.strip().str.title()
df['Composition'] = df['Composition'].str.strip().str.capitalize()
df['Uses'] = df['Uses'].str.strip().str.capitalize()
df['Side_effects'] = df['Side_effects'].str.strip().str.capitalize()
df['Manufacturer'] = df['Manufacturer'].str.strip().str.title()
```

```python
#vérification  si les pourcentages Excellent Review %, Average Review %, Poor Review % ne dépassent pas 100%
df[(df['Excellent Review %'] > 100) | (df['Average Review %'] > 100) | (df['Poor Review %'] > 100)]
```

```python
#Supprimer les lignes incomplète
df = df.dropna()
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df['Excellent Review %'], bins=30, alpha=0.7, label='Excellent Review %')
plt.hist(df['Average Review %'], bins=30, alpha=0.7, label='Average Review %')
plt.hist(df['Poor Review %'], bins=30, alpha=0.7, label='Poor Review %')
plt.xlabel('Pourcentage')
plt.ylabel('Nombre de médicaments')
plt.title('Distribution des avis sur les médicaments')
plt.legend()
plt.show()
```

```python
#voir les labo les plus present
import seaborn as sns

plt.figure(figsize=(12,6))
top_manufacturers = df['Manufacturer'].value_counts().head(10)
sns.barplot(x=top_manufacturers.index, y=top_manufacturers.values, palette="viridis")
plt.xticks(rotation=45)
plt.ylabel("Nombre de médicaments")
plt.title("Top 10 des fabricants dans le dataset")
plt.show()
```

```python
# Comparaison des pourcentages d’avis moyens par médicament les 10 plus populaires
top_meds = df.nlargest(10, 'Excellent Review %')[['Medicine Name', 'Excellent Review %', 'Average Review %', 'Poor Review %']]
top_meds.set_index('Medicine Name')[['Excellent Review %', 'Average Review %', 'Poor Review %']].plot(kind='bar', figsize=(12,6))
plt.ylabel('Pourcentage')
plt.title("Top 10 des médicaments avec les meilleurs avis")
plt.xticks(rotation=45)
plt.show()

```

```python
avg_excellent = df['Excellent Review %'].mean()
avg_average = df['Average Review %'].mean()
avg_poor = df['Poor Review %'].mean()

plt.figure(figsize=(7,7))
plt.pie([avg_excellent, avg_average, avg_poor],
        labels=['Excellent', 'Average', 'Poor'],
        autopct='%1.1f%%',
        startangle=140,
        colors=['#2ecc71', '#f1c40f', '#e74c3c'])
plt.title('Répartition moyenne des avis sur les médicaments')
plt.show()
```

```python
# si certaines descriptions sont plus complètes que d'autres
df['Uses_length'] = df['Uses'].apply(lambda x: len(str(x).split()))
plt.figure(figsize=(10,6))
sns.histplot(df['Uses_length'], kde=True, color="purple")
plt.title("Distribution de la longueur des descriptions d'utilisations")
plt.xlabel("Nombre de mots")
plt.show()
```

```python
#Top 5 des medicaments avec le pire avis
worst_meds = df.nsmallest(5, 'Excellent Review %')[['Medicine Name', 'Excellent Review %']]
print(worst_meds)
```
